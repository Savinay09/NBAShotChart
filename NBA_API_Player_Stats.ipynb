##Cleaning up dataframe

```
#convert lists to dfs
game_logs1 = pd.concat(game_logs1, ignore_index=True)
game_logs2  =  pd.concat(game_logs2, ignore_index=True)
game_logs3  =  pd.concat(game_logs3, ignore_index=True)
game_logs4  =  pd.concat(game_logs4, ignore_index=True)
game_logs5  =  pd.concat(game_logs5, ignore_index=True)
game_logs6  =  pd.concat(game_logs6, ignore_index=True)
game_logs7  =  pd.concat(game_logs7, ignore_index=True)
game_logs8  =  pd.concat(game_logs8, ignore_index=True)
game_logs9  =  pd.concat(game_logs9, ignore_index=True)
game_logs10  =  pd.concat(game_logs10, ignore_index=True)
game_logs11  =  pd.concat(game_logs11, ignore_index=True)
game_logs12  =  pd.concat(game_logs12, ignore_index=True)
game_logs13  =  pd.concat(game_logs13, ignore_index=True)
```

```
#combine dataframes
player_data = pd.concat([game_logs1,game_logs2,game_logs3,game_logs4,game_logs5,
                         game_logs6,game_logs7,game_logs8,game_logs9,game_logs10,
                         game_logs11,game_logs12,game_logs13], ignore_index=True)
```

```
player_data.shape
```

```
player_data["Player Name"]=player_data.Player_ID.map(player_dict)
```

```
player_data
```

```
#splitting up player and opposing team columns
player_data["Opponent"]=player_data.MATCHUP.str.split(expand=True)[0]
player_data["Team"]=player_data.MATCHUP.str.split(expand=True)[2]
```

```
player_data.MATCHUP.str.split(expand=True)[0]
```

```
player_data
```

```
#Fix abbreviations for old teams to match abbreviations for today's teams
player_data['Team'] = player_data['Team'].replace({'NJN':'BKN','CHA':'CHH','VAN':'MEM','NOH':'NOP','NOK':'NOP','SEA':'OKC'})
player_data['Opponent'] = player_data['Opponent'].replace({'NJN':'BKN','CHA':'CHH','VAN':'MEM','NOH':'NOP','NOK':'NOP','SEA':'OKC'})
```

```
player_data
```

```
#create season dict from season_id
seasons_dict = {'21995':'1995-1996',
                '21996':'1996-1997',
                '21997':'1997-1998',
                '21998':'1998-1999',
                '21999':'1999-2000',
                '22000':'2000-2001',
                '22001':'2001-2002',
                '22002':'2002-2003',
                '22003':'2003-2004',
                '22004':'2004-2005',
                '22005':'2005-2006',
                '22006':'2006-2007',
                '22007':'2007-2008',
                '22008':'2008-2009',
                '22009':'2009-2010',
                '22010':'2010-2011',
                '22011':'2011-2012',
                '22012':'2012-2013',
                '22013':'2013-2014',
                '22014':'2014-2015',
                '22015':'2015-2016',
                '22016':'2016-2017',
                '22017':'2017-2018',
                '22018':'2018-2019',
                '22019':'2019-2020',
                '22020':'2020-2021',
                '22021':'2021-2022',
                '22022':'2022-2023'
                }

#define Season column
player_data['SEASON']=player_data.SEASON_ID.map(seasons_dict)
```
